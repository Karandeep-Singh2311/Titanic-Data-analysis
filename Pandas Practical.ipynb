### DataFrame

```python
import pandas as pd
```

```python
df1 = pd.read_csv("titanic.csv")
df1
```

```python
df1.head(2)
```

```python
df1['fare'] = df1.fare.astype('int')
df1.head()
```

```python
def func(x):
    if x >= 35:
        return 'adult'
    elif x<35:
        return 'not adult'
    else:
        pass
df1['age'] = df1.age.map(func)
```

```python
df1.head()
```

```python
a = df1['sex']
b = df1['fare']
def discount(x,y):
    new = []
    for i,j in zip(a,b):
        if i == 'male':
            new.append(j - 10/100*j)
        elif i == 'female':
            new.append(j - 20/100*j)
        else:
            continue
    return new

df1['new_fare'] = discount(a,b)
df1.head()
```

```python
def placed(x):
    if x> 40:
        return "true"
    else:
        return "false"

df1['fit-to-placement'] = df1['fare'].map(placed)
df1.head()
```

```python
df1.head()
```

```python
def func(x):
    fare1 = []
    fare2 = []
    for i in x:
        a = str(i).split('.')
        fare1.append(int(a[0]))
        fare2.append(int(a[-1]))
    return fare1 , fare2

a,b = func(df1['fare'])
df1['fare1'] = a
df1['fare2'] = b
df1.head()
```

```python
a = df1["sex"]
b = df1["fare"]
c = df1["alive"]
def func(x,y,z):
    new = []
    for i,j, k in zip(x,y,z):
        if i=="female" and j>=50 and k=='yes':
            new.append(True)
        else:
            new.append(False)
    return new

df1["Female_Update"] = func(a,b,c)
```
